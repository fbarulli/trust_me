```python
import pandas as pd

# Assuming df is your DataFrame
def get_sentiment_and_compare(row):
    # Get the highest sentiment from RoBERTa
    roberta_scores = [row['roberta_neg'], row['roberta_neu'], row['roberta_pos']]
    sentiments = ['negative', 'neutral', 'positive']
    
    # Get the index of the highest sentiment
    highest_score_index = roberta_scores.index(max(roberta_scores))
    predicted_sentiment = sentiments[highest_score_index]
    
    # Compare the predicted sentiment with the actual rating
    # Assuming the rating is 1 for negative, 3 for neutral, and 5 for positive
    if predicted_sentiment == 'negative' and row['rating'] <= 2:
        match = True
    elif predicted_sentiment == 'neutral' and row['rating'] == 3:
        match = True
    elif predicted_sentiment == 'positive' and row['rating'] >= 4:
        match = True
    else:
        match = False
    
    return predicted_sentiment, match

# Apply the function to each row and create new columns for the predicted sentiment and comparison
df[['predicted_sentiment', 'match']] = df.apply(get_sentiment_and_compare, axis=1, result_type='expand')

# Display the updated DataFrame
print(df[['review', 'predicted_sentiment', 'rating', 'match']].head())
```
