```python
import pandas as pd
dataframe_2 = pd.read_csv("final_output.csv")

df_cleaned = dataframe_2.dropna(subset=["Company Name", "Trust Score", "Number of Reviews"])
display(df_cleaned)
```

```python
import matplotlib.pyplot as plt

# 1. Distribution of Trust Scores
plt.figure(figsize=(8, 5))
plt.hist(df_cleaned['Trust Score'], bins=30, edgecolor='black')
plt.title('Distribution of Trust Scores')
plt.xlabel('Trust Score')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()
```

```python
# 2. Number of Reviews per Company (Top 10 Companies by Number of Reviews)
top_10_reviews = df_cleaned.nlargest(10, 'Number of Reviews')
plt.figure(figsize=(10, 6))
plt.barh(top_10_reviews['Company Name'], top_10_reviews['Number of Reviews'], color='skyblue')
plt.title('Top 10 Companies by Number of Reviews')
plt.xlabel('Number of Reviews')
plt.ylabel('Company Name')
plt.gca().invert_yaxis()
plt.show()
```

```python
# 3. Scatter Plot: Trust Score vs. Number of Reviews
plt.figure(figsize=(8, 5))
plt.scatter(df_cleaned['Number of Reviews'], df_cleaned['Trust Score'], alpha=0.7)
plt.title('Trust Score vs. Number of Reviews')
plt.xlabel('Number of Reviews')
plt.ylabel('Trust Score')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns

numeric_df = df_cleaned.select_dtypes(include=['number'])

# Plot the heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```
